```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats
import statistics as stat
import time


def import_data(file):
    # Read into data frame using pandas
    data = pd.read_csv(file,
                       sep=',',
                       header=0,
                       parse_dates=True)
    return data
```

```python
def multi_boxplot(data):
    col_labels = []
    plots = []
    for column in data:
        col_labels.append(column)
        plots.append(data[column])
    plt.boxplot(plots, labels=col_labels, meanline=True, vert=0)
    plt.title('Multi-BoxPlot')
    plt.show()

```
